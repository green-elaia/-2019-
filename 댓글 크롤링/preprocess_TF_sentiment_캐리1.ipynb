```
okt = Okt()
with open('/gdrive/My Drive/Colab Notebooks/graduation_paper/selected_words.json', encoding = 'UTF-8') as f:
        selected_words = json.load(f)
model = models.load_model('/gdrive/My Drive/Colab Notebooks/graduation_paper/second_model.h5')


def tokenize(doc):
    return ['/'.join(t) for t in okt.pos(doc, norm=True, stem=True)]

def term_frequency(doc):
    return [doc.count(word) for word in selected_words]

def predict_sentiment(review):
    token = tokenize(review)
    tf = term_frequency(token)
    data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
    score = float(model.predict(data))
    if(score > 0.5):
        return 1
    else:
        return 0
```

```
def preprocess_keyword_csv(keywordName):
    global keyword_comment
    keyword_comment = []
    length = 0
    
    with open("/gdrive/My Drive/Colab Notebooks/graduation_paper/union_keyword_url/" + keywordName + ".csv", 'r', encoding = 'cp949') as p_csv:
        length = len(list(csv.reader(p_csv)))

    with open("/gdrive/My Drive/Colab Notebooks/graduation_paper/union_keyword_url/" + keywordName + ".csv", 'r', encoding = 'cp949') as p_csv:
        print("yeah")
        keyword_data = csv.reader(p_csv)
        print("leng : ", length)
        i = 0

        for row in keyword_data:
          url = row[0]
          flag = False
          print(keywordName, " : ", i, " / ", length)
          i += 1
          
          for pp_data in pp_comment:
            if url in pp_data[0]:
              keyword_comment.append([pp_data[0],pp_data[1], pp_data[2],pp_data[3], pp_data[4], predict_sentiment(pp_data[3])])
              #before_url = pp_data[0]
              flag = True
              continue
            if flag and url not in pp_data[0]:
              break
```
